# Exploratory Data Analysis(EDA) Using Python-Part 2

```python
# Importing all the library files required

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

from scipy import stats
```

```python
df=pd.read_csv('train.csv')   # reading the file from the location
```

```python
df.head()    # shows the first five rows of the dataset
```

```python
df.hist(figsize=(20,20))  
# Histogram of each column having Numerical Values
# ApplicantIncome and CoapplicantIncome columns are right skewed
```

```python
sns.pairplot(df, hue="Loan_Status") # used to plot multiple pairwise bivariate distributions in a dataset
```

```python
cor=df.corr()
sns.heatmap(cor,vmin=-1,vmax=1,annot=True,linewidths=.5,cmap='bwr')
# heatmap is used to plot the correlation between different numerical columns
```
